Muestra un DataFrame con los datos de la siguiente tabla:

```
  Mes     Ingresos     Gastos
  Enero   21500        17000
  Febrero 17600        15400
  Marzo   25000        19200
  Abril   34800        22500
```



````python
datos = {'Mes':['Enero', 'Febrero', 'Marzo', 'Abril'], 'Ingreso': ['21500', '17600', '25000', '34800'], 'Gastos': ['17000', '15400', '19200', '22500']}
df = pd.DataFrame(datos)
df
````

:Escribe una funcion que reciba el DataFrame del ejercicio anterior y devuelva el beneficio total de cada mes (Ingresos - Gastos).


````python
def beneficio(datos):
  # Access columns using bracket notation and correct column names
  datos['beneficio'] = datos['Ingreso'].astype(int) - datos['Gastos'].astype(int)
  return datos
````

````python
beneficio(df)
````

````python
titanic = pd.read_csv('titanic.csv', index_col=0)
titanic.head()
````

````python
#2. Mostrar Dimensiones
print("Dimensiones: ", titanic.shape)
print("Numero de datos: ", titanic.size)
print("Nombres de las columnas: ", titanic.columns)
print("Nombres de las filas: ", titanic.index)
print("Tipos de datos: \n", titanic.dtypes)
print("Primeras 5 filas: \n", titanic.head())
print("Ultimas 5 filas: \n", titanic.tail())
````

````python
#3 Pasajero 121
titanic.loc[121]
````

````python
#4 Filas impares
titanic.iloc[range(0,titanic.shape[0],2)]
````

````python
#5 Supervivientes
titanic['Survived'].value_counts()/titanic['Survived'].count()*100
````

````python
#6. Segunda clase
#titanic[titanic['Pclass']==2].sort_values('Name')
pasajeros_segunda=titanic[titanic['Pclass']==2]['Name'].sort_values()
pasajeros_segunda.head(5)
````

````python
#7.Eliminar nulos Edad
#titanic.dropna(subset=['Age'])
titanic=titanic.dropna(subset=['Age'])
titanic.head()
````

````python
#8.Muertos por clase
titanic.groupby('Pclass')['Survived'].value_counts(normalize= True)*100
````

````python
#9. Edad media por clase
titanic.groupby(['Pclass', 'Sex'])['Age'].mean()
````

````python
#10. Menor de edad
titanic['Menor']=titanic['Age']<18
titanic['Menor'].head()
#titanic.head()
````

````python
#11. Mostrar % de Menores y Mayores de Edad
titanic.groupby(['Pclass', 'Menor'])['Survived'].value_counts(normalize= True)*100
````
